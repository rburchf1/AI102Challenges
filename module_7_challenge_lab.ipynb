```python
# ---------------------------------------------------------
# ADVANCED SENTIMENT ANALYSIS WITH LSTM (CHALLENGE LAB)
# ---------------------------------------------------------
# Students must complete the missing sections marked as TODO.
# This version includes:
# - Two LSTM layers
# - Dropout for regularization
# - Multiple evaluation metrics
# ---------------------------------------------------------

import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# ---------------------------------------------------------
# 1. TEXT DATASET AND LABELS
# ---------------------------------------------------------
corpus = [
    "I love this product! It's amazing.",
    "This is the worst service I have ever experienced.",
    "I am very happy with my purchase.",
    "I am disappointed with the quality of this item.",
    "The experience was fantastic and enjoyable.",
    "I will never buy from this company again."
]

labels = [1, 0, 1, 0, 1, 0]   # 1 = Positive, 0 = Negative


# ---------------------------------------------------------
# 2. TOKENIZATION AND PADDING
# ---------------------------------------------------------

# TODO 1: Create a Tokenizer that keeps the top 6000 words
tokenizer = Tokenizer(num_words=6000)

# TODO 2: Fit the tokenizer on the corpus
tokenizer.fit_on_texts(corpus)

# TODO 3: Convert the text into sequences
X = tokenizer.texts_to_sequences(corpus)

# TODO 4: Pad or truncate all sequences to length 12
X = pad_sequences(X, maxlen=12)


# ---------------------------------------------------------
# 3. TRAIN / TEST SPLIT
# ---------------------------------------------------------

# TODO 5: Split into training and test sets (30% test data)
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.30, random_state=42
)


# ---------------------------------------------------------
# 4. BUILD THE LSTM MODEL
# ---------------------------------------------------------

model = Sequential()

# TODO 6: Add an Embedding layer
# input_dim should be the vocabulary size
# output_dim should be 64
# input_length should match your padding length
model.add(Embedding(input_dim=6000, output_dim=64, input_length=12))

# TODO 7: Add the first LSTM layer with 128 units AND return_sequences=True
model.add(LSTM(128, return_sequences=True))

# TODO 8: Add a Dropout layer with rate 0.3
model.add(Dropout(0.3))

# TODO 9: Add a second LSTM layer with 64 units
model.add(LSTM(64))

# TODO 10: Add another Dropout layer with rate 0.3
model.add(Dropout(0.3))

# TODO 11: Add a Dense hidden layer with 16 units and ReLU activation
model.add(Dense(16, activation='relu'))

# TODO 12: Add the final output Dense layer (1 unit, sigmoid activation)
model.add(Dense(1, activation='sigmoid'))


# ---------------------------------------------------------
# 5. COMPILE THE MODEL
# ---------------------------------------------------------

# TODO 13:
# Compile the model using:
# - optimizer = "adam"
# - loss = "binary_crossentropy"
# - metrics = ["accuracy"]
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)


# ---------------------------------------------------------
# 6. TRAIN THE MODEL
# ---------------------------------------------------------

# TODO 14:
# Train the model for 8 epochs using validation data
model.fit(
    X_train,
    np.array(y_train),
    epochs=8,
    verbose=1,
    validation_data=(X_test, np.array(y_test))
)


# ---------------------------------------------------------
# 7. MODEL EVALUATION WITH MULTIPLE METRICS
# ---------------------------------------------------------

# TODO 15: Generate predictions on the test set
y_pred = model.predict(X_test)

# TODO 16: Convert probabilities into class labels (0 or 1)
y_pred_classes = (y_pred > 0.5).astype(int).flatten()

accuracy  = accuracy_score(y_test, y_pred_classes)
precision = precision_score(y_test, y_pred_classes)
recall    = recall_score(y_test, y_pred_classes)
f1        = f1_score(y_test, y_pred_classes)

print("Model Performance:")
print(f"Accuracy : {accuracy}")
print(f"Precision: {precision}")
```
