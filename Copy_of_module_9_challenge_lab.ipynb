```python
# ----------------------------------------------------------
# CHALLENGE LAB: TEXT SUMMARIZATION WITH BART (TRANSFORMERS)
# ----------------------------------------------------------
# In this challenge, you will:
# 1) Load a pre-trained BART model and tokenizer.
# 2) Summarize MULTIPLE input texts in a loop.
# 3) Experiment with different generation settings.
# 4) Compute a simple compression ratio for each summary.
# ----------------------------------------------------------

from transformers import BartForConditionalGeneration, BartTokenizer

# ----------------------------------------------------------
# 1. LOAD THE PRE-TRAINED BART MODEL AND TOKENIZER
# ----------------------------------------------------------

# TODO 1:
# Set the model name to "facebook/bart-large-cnn"
model_name = "facebook/bart-large-cnn"

# TODO 2:
# Load the model and tokenizer using from_pretrained(...)
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

# ----------------------------------------------------------
# 2. SAMPLE TEXTS TO SUMMARIZE
# ----------------------------------------------------------
# We will summarize more than one text.
# Feel free to change or add your own long paragraphs.

texts = [
    """Natural language processing (NLP) is a subfield of linguistics, computer science, and artificial intelligence
concerned with the interactions between computers and human language. It focuses on how to program computers to process
and analyze large amounts of natural language data. Common NLP tasks include speech recognition, natural language understanding,
and natural language generation.""",

    """Machine learning is a subset of artificial intelligence that focuses on building systems that can learn from data.
Instead of being explicitly programmed with rules, these systems use algorithms to find patterns in data and improve their
performance over time. Applications of machine learning include recommendation systems, image recognition, and fraud detection."""
]

# ----------------------------------------------------------
# 3. SET GENERATION PARAMETERS
# ----------------------------------------------------------
# We will use variables so it is easy to experiment with different values.

# TODO 3: Choose values for these parameters:
max_summary_length=150   # e.g., 80
min_summary_length=40   # e.g., 30
num_beams=4            # e.g., 4
length_penalty=2       # e.g., 2.0

# ----------------------------------------------------------
# 4. SUMMARIZE EACH TEXT
# ----------------------------------------------------------

for idx, text in enumerate(texts):
    print("=" * 60)
    print(f"ORIGINAL TEXT #{idx + 1}:")
    print(text)
    print("\n--- Generating summary... ---\n")

    # TODO 4:
    # Tokenize the input text.
    # Use tokenizer.encode with:
    #   - "summarize: " + text   (prefix)
    #   - return_tensors="pt"
    #   - max_length=512
    #   - truncation=True
    inputs = tokenizer.encode(
        "summarize: " + text,
        return_tensors="pt",
        max_length=512,
        truncation=True
    )

    # TODO 5:
    # Generate the summary using model.generate with:
    #   - input_ids = inputs
    #   - max_length = max_summary_length
    #   - min_length = min_summary_length
    #   - num_beams = num_beams
    #   - length_penalty = length_penalty
    #   - early_stopping = True
    summary_ids = model.generate(
        inputs,
        max_length=150,
        min_length=40,
        num_beams=4,
        length_penalty=2.0,
        early_stopping=True
    )

    # TODO 6:
    # Decode the first summary in summary_ids back to text.
    # Use tokenizer.decode with skip_special_tokens=True
    summary = tokenizer.decode(
        summary_ids[0],
        skip_special_tokens=True
    )

    print("SUMMARY:")
    print(summary)

    # ------------------------------------------------------
    # 5. COMPUTE A SIMPLE COMPRESSION RATIO
    # ------------------------------------------------------
    # We'll compare the number of characters in the original text
    # to the number of characters in the summary.

    # TODO 7:
    original_length = len(text)
    summary_length = len(summary)

    # To avoid division by zero, make sure summary_length is not zero.
    if summary_length > 0:
        compression_ratio = summary_length / original_length
    else:
        compression_ratio = 0.0

    print("\n--- Compression Info ---")
    print(f"Original length (chars): {original_length}")
    print(f"Summary length  (chars): {summary_length}")
    print(f"Compression ratio       : {compression_ratio:.2f}")  # < 1.0 = shorter summary


```
