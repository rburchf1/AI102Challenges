# 🧪 Module-05: Challenge Lab

🧪 Challenge Lab: Building a Custom N-gram Analyzer

🎯 Learning Goals
* Loads and preprocesses a multi-paragraph text
* Removes punctuation and English stopwords
* Generates N-grams for ANY value of N provided by the user
* Counts and ranks the most common N-grams
* Compares N-gram frequencies across two different sections of text

🧱 Students must complete the missing sections marked as TODO

✅  When they complete and run the code, they should see:
* Top 10 3-grams
* Most common bigrams in section 1
* Most common bigrams in section 2

```python
# --------------------------------------------------------------
# ADVANCED N-GRAM LAB — Custom Analyzer and Section Comparison
# --------------------------------------------------------------

from nltk import ngrams
from collections import Counter
import nltk, string

# Download tokenizers and stopwords
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))

# --------------------------------------------------------------
# 1. LOAD MULTI-PARAGRAPH TEXT
# --------------------------------------------------------------

text = """
Natural language technologies have rapidly evolved over the past decade.
Businesses now rely on NLP to automate emails, analyze customer feedback,
and support decision making.

Deep learning architectures such as Transformers have revolutionized
machine translation, summarization, and conversational AI systems.
"""

# --------------------------------------------------------------
# 2. PREPROCESSING FUNCTION
# --------------------------------------------------------------

def preprocess(text):
    # TODO: convert text to lowercase
    text = text.lower()

    # TODO: remove punctuation using translate()
    text = text.translate(str.maketrans("", "", string.punctuation))

    # tokenize into words
    tokens = nltk.word_tokenize(text)

    # TODO: remove stopwords
    tokens = [t for t in tokens if t.lower() not in stop_words]

    return tokens

tokens = preprocess(text)

# --------------------------------------------------------------
# 3. FUNCTION TO CREATE DYNAMIC N-GRAMS
# --------------------------------------------------------------

def generate_ngrams(tokens, n):
    return [" ".join(g) for g in ngrams(tokens, n)]

# --------------------------------------------------------------
# 4. USER-DEFINED N (HARDER PART)
# --------------------------------------------------------------

# TODO: Ask the user for a number N (1–5 recommended)
# Convert it to an integer
n = int( input("Enter N for N-grams (e.g., 1, 2, 3): ") )

# TODO: generate N-grams using the user's chosen N
generated = generate_ngrams(tokens, n)

# --------------------------------------------------------------
# 5. FREQUENCY ANALYSIS
# --------------------------------------------------------------

# TODO: compute frequency counts
freqs = Counter(generated)

print(f"\nTop 10 {n}-grams:")
print(freqs.most_common(10))

# --------------------------------------------------------------
# 6. SECTION-BASED ANALYSIS (Harder Challenge)
# --------------------------------------------------------------

# Break text into two sections
section1, section2 = text.split("\n\n")

tokens1 = preprocess(section1)
tokens2 = preprocess(section2)

# TODO: generate bigrams for each section
bigrams1 = generate_ngrams(tokens1, 2)
bigrams2 = generate_ngrams(tokens2, 2)

# TODO: count frequency of bigrams for each section
freq1 = Counter(bigrams1)
freq2 = Counter(bigrams2)

print("\nMost common bigrams in section 1:")
print(freq1.most_common(5))

print("\nMost common bigrams in section 2:")
print(freq2.most_common(5))
```
